### matplotlib 차트 유형

1. histogram : 연속형 값의 도수(빈도수/개수) 파악
2. bar plot : 범주형 값의 개수, 또는 다른 연속값을 파악
3. scatter plot : 산포도(산점도) 두 연속값들의 분포 파악
4. line plot : 시계열데이터(연속값) 변화 파악
5. box plot : 연속형 값의 사분위, 이상치 파악
6. pie chart : 전체 범위 중의 비율

##### Histogram

- 연속형 값을 일정 구간을 나누어 구간별 데이터포인트의 빈도를 시각화
    - x축 : 연속형 데이터
    - y축 : 구간별 데이터포인트의 개수 (합)

```python
scores = [100, 95, 97, 98, 99, 90, 80, 81, 84, 88, 70, 78, 79, 70, 65, 66, 61, 64, 100, 89]

bins = [60, 70, 80, 90, 100]
plt.hist(scores, bins=bins, edgecolor='black',rwidth=0.8)
# bins = 나뉠 구간의 갯수 (기본값 = 10)
# edgecolor : 히스토그램 막대의 경계선 색상
# rwidth : 0~1 사이의 값으로 상대적 너비 조정

plt.xlabel('점수')
plt.ylabel('인원수')
plt.title('점수별 인원도수표')
plt.show()
```

```python
plt.hist(scores,edgecolor='black',rwidth=0.95,cumulative=True)
# cumulatve : 누적 값
plt.show()
```

##### Bar Plot

- 막대 차트
- 범주형 데이터와 연속형 데이터(막대의 길이)의 관계를 시각화
    - x축 : 범주형 데이터 카테고리
    - y축 : 연속형 데이터 (총합, 평균)

```python
# 월별 매출 데이터
df=pd.DataFrame({
    'month' : ['1월','2월','3월','4월','5월','6월'],
    'sales' : [1300,1600,1700,1650,1900,2300]
})

# zorder : z축 깊이 설정 (값이 클수록 앞으로 나옴)
plt.bar(df['month'],df['sales'],zorder=10)

for i,val in enumerate(df['sales']):
    # x축 위치, y축 위치, 텍스트, 텍스트 정렬(위치)
    plt.text(i,val+100,str(val),ha='center')

plt.xlabel('월')
plt.ylabel('매출(단위 : 천만원)')
plt.grid(zorder=1)

plt.show()
```

```python
# 수평 막대그래프
plt.barh(df['month'],df['sales'])

plt.ylabel('월')
plt.xlabel('매출(단위 : 천만원)')
plt.show()
```

##### Line plot

##### Scatter Plot

- 둑 축의 변수 관계를 시각화
- 두 변수 간 상관관계, 클러스터링, 이상치 파악에 도움

```python
df = pd.DataFrame({
    'ad_budget': np.arange(100, 1001, 100), 
    'sales': [30, 45, 50, 70, 85, 90, 105, 130, 150, 180],
    'visitor': [100, 140, 200, 210, 260, 290, 300, 350, 440, 600],
    'colors': [25, 50, 70, 90, 100, 200, 235, 300, 450, 700]
})

# x축, y축, s=점의 크기, c=점의 색상(컬러 값으로 연속 데이터 표현)
plt.scatter(df['ad_budget'],df['sales'],s=df['visitor'],c=df['colors'])

plt.colorbar()

plt.xlabel('광고비')
plt.ylabel('광고비')
plt.show()
```

##### Pie Chart

- 전체에서 각 부분이 기여하는 비율을 시각화

```python
df = pd.DataFrame({
    'label': ['애플', '삼성', '샤오미', 'LG'],
    'market_share': [33, 32, 10, 25], 
    'sales': [1345, 234, 545, 454],
    'explode': [0, 0.1, 0, 0]
})

plt.pie(
    df['market_share'],
    labels=df['label'],
    autopct='%.1f%%',
    startangle=90,
    explode=df['explode'],
    colors=['skyblue','lightgreen','silver','gold']
)

plt.title('기업별 시장 점유율')
plt.show()
```

```python
plt.pie(
    df['sales'],
    labels=df['label'],
    autopct='%.1f%%',
    startangle=90,
    explode=[0.05,0.05,.05,.05],
    wedgeprops={'width': .6}
)
plt.text(0,0,'기업별 판매 비율',ha='center',va='center')

plt.show()
```

##### Box Plot

- 연속형 데이터의 분포를 사분위수로 시각화
- 이상치 outlier를 도출하는 데에 효과적
    - x축 : 이산값(범주형) 데이터
    - y축 : 연속형 데이터

- 표현되는 값들
    - max, Q3, Q2, Q1, min : 최대값, 3분위수(75%), 2분위수(50%, 중위값), 1분위수(25%), 최소값
    - IQR (Inter-Quatile Range) : 사분범위

```python
# 반별 시험 성적 데이터 (2차원 배열)
class_scores = [
    [85, 88, 90, 92, 93, 87, 95, 91, 84, 86],  # Class A
    [78, 82, 85, 80, 79, 84, 81, 83, 77, 76],  # Class B
    [92, 95, 96, 94, 90, 97, 93, 91, 98, 99],  # Class C
    [70, 72, 68, 75, 74, 69, 73, 71, 66, 67]   # Class D
]

plt.boxplot(class_scores)

plt.xlabel('반')
plt.ylabel('점수')
plt.show()
```

```python
data=class_scores[0]

print('최솟값:',np.min(data))
print('사분위수(Q1):',np.percentile(data,75))
print('사분위수(Q2):',np.percentile(data,50),np.median(data))
print('사분위수(Q3):',np.percentile(data,25))
print('최댓값:',np.max(data))
```

```python
weird_data = [
    [100, 120, 130, 110, 115, 105, 125,200],   # 이상치 200
    [150, 160, 170, 155, 165, 158],
    [80, 85, 90, 75, 92, 87,300],   # 이상치 300
    [200, 220, 210, 205, 215, 230,50]  # 이상치 50
]

# whis 기본값 : 1.5
# [참고] 수염의 길이 계산 공식 : [Q1 - whis * IQR, Q3 + whis * IQR]
# [참고] IQR = Q3 - Q1
plt.boxplot(weird_data,whis=5)
plt.xticks([1,2,3,4,],['상품1','상품2','상품3','상품4'])
plt.xlabel('상품명')
plt.ylabel('싯가')
plt.show()
```
