### 통계 관련 numpy

```python
import numpy as np
```

##### 중심 경향 측도

```python
height=np.array([180, 181, 166, 179, 183, 178, 179, 165, 160, 160, 185, 184, 174, 160, 162, 184])
```

```python
# 평균값
print(np.mean(height))
```

```python
# 중위값
print(np.sort(height))
print(np.median(height))
```

```python
# 최빈값
values,count=np.unique(height,return_counts=True)
print(values,count)

max_count=np.max(count)
print(count==max_count)
print(values[count==max_count])
```

---

### 산포 측도

```python
# 분산 variance
# (값 - 평균)^2의 평균

var1=np.mean(np.square(height -np.mean(height)))
var2=np.var(height)

print(var1)
print(var2)
```

```python
# 표준편차 standard deviation 
# 분산의 제곱근
std1=np.sqrt(var1)
std2=np.std(height)
print(std1)
print(std2)
```

```python
# 범위
print(np.max(height)-np.min(height))
```

```python
import matplotlib.pyplot as plt
```

```python
# 정규분포
# np.random(shape)
# 평균이 0, 표주편차가 1인 난수 출력 함수

np.random.randn()
```

```python
arr=np.random.randn(1000)

plt.hist(arr,bins=100) # 히스토그램
plt.show() # 출력
```

```python
# np.random.normal(loc,scale,size)
# - loc : 평균
# - scale : 표준 편차
# - size : 갯수 or shape

# arr=np.random.normal(10,1,1000)
arr=np.random.normal(10,1,(10,100))
print(arr)

plt.hist(arr,bins=10)
plt.show()
```

```python
# np.random.rand(num)
# - 0.0 이상 ~ 1.0 미만 수에서 균등하게 난수 추출
# - num : 갯수 or shape
arr=np.random.rand(1000)
arr=np.random.rand(2,3)

plt.hist(arr)
plt.show()
```
