````python
arr = np.array([10, 20, 30, 30, 40, 50, 60])

print(np.mean(arr))
print(np.median(arr))

values,count=np.unique(arr,return_counts=True)
max_count=np.max(count)
print(values[count==max_count])

print(np.var(arr))
print(np.std(arr))
print(np.max(arr)-np.min(arr))
````

### [**실습 3] 평균 성적 구하기**

<aside>
❓아래와 같이 학생별 과목 성적 데이터가 주어졌을 때, 각 학생의 전체 평균 성적(수학, 영어, 과학)을 계산한 후, 이 평균값에 대한 분산과 표준편차를 구하세요.

</aside>

```python
# 학생별 과목 성적 데이터
scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]
```

````python
scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]

scores_math_np=np.array(scores_math)
scores_english_np=np.array(scores_english)
scores_science_np=np.array(scores_science)

total=scores_math_np+scores_english_np+scores_science_np

total_avg=total//3
print(total_avg)
print(np.var(total_avg))
print(np.std(total_avg))
````
